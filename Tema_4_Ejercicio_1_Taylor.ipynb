# Ejercicio 1. 

## Dada la función $f(x)=\cos{x}+\sin{x}$, se pide: 
<p><strong>(i) Halla su polinomio de Taylor de orden 5 centrado en $x_0=0$.

(ii) Representa gráficamente la función $f(x)$ y el polinomio anterior.

(iii) Calcula una aproximación de $\cos{(0.1)}+\sin{(0.1)}$ utilizando el polinomio anterior.

(iv) Da una cota superior del error absoluto de la aproximación hallada en (iii).</strong></p>

SOLUCIÓN:

<p><strong>Apartado (i)

Halla su polinomio de Taylor de orden 5 centrado en $x_0=0$</strong></p>

Definimos la función:

```python
f(x)=cos(x)+sin(x)
f
```

<p>Podemos calcular el polinomio pedido mediante el comando "taylor":</p>

```python
P(x)=taylor(f,x,0,5)
P(x)
```

<p><strong>Apartado (ii)

Representa gráficamente la función $f(x)$ y el polinomio anterior.</strong></p>

Hacemos una representación gráfica en el intervalo $[-3,3]$:

```python
plot(f,(x,-3,3),legend_label='$y=f(x)$')+plot(P,(x,-3,3),legend_label='$y=P(x)$',color='red')
```

<p>Se observa que $P(x)$ es una buena aproximación de $f(x)$ para valores de $x$ en el intervalo [-1,1].</p>
<p>Podemos verlo dibujando la función de "error" $|f(x)-P(x)|$ en el intervalo [-1.5,1.5]:</p>

```python
plot(abs(f(x)-P(x)),(x,-1.5,1.5),legend_label='$y=|f(x)-P(x)|$',color='green')
```

<p><strong>Apartado (iii)

Calcula una aproximación de $\cos{(0.1)}+\sin{(0.1)}$ utilizando el polinomio anterior.</strong></p>

La aproximación pedida es:

```python
aprox=P(0.1); aprox
```

<p><strong>Apartado (iv)
    
Da una cota superior del error absoluto de la aproximación hallada en (iii). </strong></p>
 
Una cota superior del error absoluto $|f(0.1)-P(0.1)|$ es $\dfrac{M\cdot (0.1)^6}{6!}$, siendo $M$ una constante cumpliendo
$$|f^{(6)}(t)|\leq M \quad\text{para todo }t\in [0,0.1].$$

Buscamos $M$ gráficamente:

```python
plot(abs(diff(f(x),x,6)),(x,0,0.1),gridlines='true')+plot(1.1,(x,0,0.1),color='red')
```

<p>Podemos tomar $M=1.1$ y así una cota del error será:</p>

```python
cota=1.1*10^(-6)/factorial(6); cota
```

<p>Comprobamos que, efectivamente, $|f(0.1)-P(0.1)|$ es menor que dicha cota:</p>

```python
f(0.1)
```

```python
P(0.1)
```

```python
error=abs(f(0.1)-P(0.1))
error
```

```python
error.n() < cota.n() #Comprobación 
```
